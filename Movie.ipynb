```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('netflix_titles_nov_2019.csv.zip')
```

```python
df.head()
```

```python
# show_id
# title
# director
# cast
# duration
# type
movie = df[['show_id','title','director','cast','duration','type','description','listed_in']]
```

```python
movie.info()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.head()
```

```python
movie.shape
```

```python
def remove_internal_spaces_from_names(names_string):
    if not isinstance(names_string, str) or not names_string.strip():
        return ""
    individual_names = [name.strip() for name in names_string.split(',')]

    processed_names = [name.replace(' ', '') for name in individual_names]

    return ','.join(processed_names)

movie['director'] = movie['director'].apply(remove_internal_spaces_from_names)
```

```python
movie.head()
```

```python
movie['cast'] = movie['cast'].apply(remove_internal_spaces_from_names)
```

```python
movie.head()
```

```python
movie['tags'] =  movie['description'] + movie['listed_in'] + movie['director'] + movie['cast']  + movie['duration']
```

```python
movie.head()
```

```python
nmovie = movie[['show_id','title','tags']]
```

```python
nmovie
```

```python
nmovie['tags'] = nmovie['tags'].apply(lambda x:x.lower())
```

```python
nmovie.head()
```

```python
nmovie.shape
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem (text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

        return " ".join(y)
```

```python
nmovie['tags'] = nmovie['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=4000,stop_words='english')
```

```python
vector =cv.fit_transform(nmovie['tags']).toarray()
```
