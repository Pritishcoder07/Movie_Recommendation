```python
def remove_internal_spaces_from_names(names_string):
    if not isinstance(names_string, str) or not names_string.strip():
        return ""
    individual_names = [name.strip() for name in names_string.split(',')]

    processed_names = [name.replace(' ', '') for name in individual_names]

    return ','.join(processed_names)

movie['director'] = movie['director'].apply(remove_internal_spaces_from_names)
```

```python
movie['cast'] = movie['cast'].apply(remove_internal_spaces_from_names)
```

```python
def recommend(movie_title):
    try:
        movie_index = nmovie[nmovie['title'] == movie_title].index[0]
    except IndexError:
        print(f"Movie '{movie_title}' not found in the dataset.")
        return

    distances = similarity[movie_index]
    # Get the top 5 most similar movies (excluding itself) and their distances
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    print(f"Recommendations for '{movie_title}':")
    for i in movies_list:
        print(f"  - {nmovie.iloc[i[0]].title} (Similarity: {i[1]:.4f})")
```

```python
recommend('Atlantics')
```
