````python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

````python
df= pd.read_csv('netflix_titles_nov_2019.csv.zip')
````

````python
df.head()
````

````python
# show_id
# title
# director
# cast
# rating
# duration
# type
movie = df[['show_id','title','director','cast','duration','type']]
````

````python
# Re-create movie DataFrame from original df to ensure correct data
movie = df[['show_id','title','director','cast','duration','type']].copy()
````

````python
# Fill NaN values in 'director' and 'cast' columns with empty strings
movie['director'] = movie['director'].fillna('')
movie['cast'] = movie['cast'].fillna('')
````

The `convert3` function should be defined in a previous cell (cell `17f97521`). If that cell was not run or was modified, please ensure it contains the correct definition:
```python
def convert3(obj):
  L = []
  if isinstance(obj, str) and obj.strip() != '':
    names = [name.strip() for name in obj.split(',')]
    for name in names:
        if len(L) < 3: # Get up to 3 names
          L.append(name)
        else:
          break
  return L
```

````python
# Apply the director processing to split names into lists
movie['director'] = movie['director'].apply(lambda x: [d.strip() for d in x.split(',')] if isinstance(x, str) and x.strip() != '' else [])
````

````python
# Apply the convert3 function to the 'cast' column to get up to 3 names
movie['cast'] = movie['cast'].apply(convert3)
````

````python
# Process 'duration' column to split into lists (e.g., ['1', 'Season'] or ['67', 'min'])
movie['duration'] = movie['duration'].apply(lambda x: x.split() if isinstance(x, str) else x)
````

````python
# Process 'type' column to split into lists (e.g., ['TV', 'Show'] or ['Movie'])
movie['type'] = movie['type'].apply(lambda x: x.split() if isinstance(x, str) else x)
````

````python
def is_empty_list(lst):
    return not lst or all(not item.strip() for item in lst)
````

````python
movie = movie[~movie['director'].apply(is_empty_list) & ~movie['cast'].apply(is_empty_list)]
````
